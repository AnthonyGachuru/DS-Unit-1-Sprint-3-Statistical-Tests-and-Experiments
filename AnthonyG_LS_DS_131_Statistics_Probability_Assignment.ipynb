```python
rep_pty = df[df.party == 'republican']
print(rep_pty.shape)
rep_pty.head()
```

```python
dem_pty = df[df.party == 'democrat']
print(dem_pty.shape)
dem_pty.head()
```

##HYPOTHESIS TESTING


```python
from scipy.stats import ttest_1samp, ttest_ind, ttest_ind_from_stats, ttest_rel


```

###1 Sample Test


```python
dem_pty['handicapped-infants'].mean()
```

####H0: There is no support for the handicapped infant bill
####Ha: There is some support for the aforemetioned bill


```python
ttest_1samp(dem_pty['handicapped-infants'], 0, nan_policy='omit')

#Due to a t-statistic of 19.826 and pvalue of 0.00000000, we reject the null
#hypothesis that there is 0 support for the aforementioned bill by democrats 
#and suggest the alternative hypothesis that there is some support for the bill
```

####H0: Support is evenly divided amongst democrats
####Ha: Support is not evenly divided

```python
ttest_1samp(dem_pty['handicapped-infants'], 0.5, nan_policy='omit')

#Due to the t-statistic of 3.43 and a pvalue of 0.0006, we reject the null 
#hypothesis that support is evenly divided 50/50 and suggest the alternative hypothesis 
#that their support is not evenly divided 50/50
```

##2 Sample Test

####H0: Both parties have similar positions on immigration
####Ha: Both parties have different positions on immigration

```python
ttest_ind(dem_pty['immigration'], rep_pty['immigration'], nan_policy='omit')

#Due to the t-statistics of 1.736 and pvalue of 0.08, we accept the null hypothesis
#that both parties have similar positions on immigration and reject the alternative 
#hypothesis
```

##Issue where Democrats support more than Republicans

####H0: Republicans support the budget policy as much as Democrats
####Ha: Support for budget policy from the Democratic party is much higher

```python
ttest_ind(dem_pty['budget'], rep_pty['budget'], nan_policy = 'omit')

#Due to the t-statistic of 23.21 and a pvalue of 0.000000, we reject the null hypothesis and accept the alternative
```

##Issue where Republicans support more than Democrats

####H0: Democrats oppose anti-satellite-ban as much as Republicans
####Ha: Republicans support the bill more

```python
ttest_ind(dem_pty['anti-satellite-ban'], rep_pty['anti-satellite-ban'], nan_policy='omit')

#Due to the t-statistic of 12.526 and a pvalue of 0.00000000, we reject the null hypothesis and suggest the alternative hypothesis
```

##Issue where the difference between republicans and democrats has p > 0.1

####H0: Support on water projects for both parties are similar
####Ha: Support is different

```python
ttest_ind(dem_pty['water-project'], rep_pty['water-project'], nan_policy='omit')

#Due to a pvalue of 0.929, we fail to reject the null hypothesis and reject the alternative hypothesis
```
